```python
import pandas as pd
import os

daily_activity_path = os.path.join(path, 'mturkfitbit_export_4.12.16-5.12.16', 'Fitabase Data 4.12.16-5.12.16', 'dailyActivity_merged.csv')
heart_rate_path = os.path.join(path, 'mturkfitbit_export_4.12.16-5.12.16', 'Fitabase Data 4.12.16-5.12.16', 'heartrate_seconds_merged.csv')

daily_activity_df = pd.read_csv(daily_activity_path)
heart_rate_df = pd.read_csv(heart_rate_path)

display(daily_activity_df.head())
display(heart_rate_df.head())
```

## Merge data

### Subtask:
Merge the two dataframes on the 'ActivityDate'.


**Reasoning**:
Convert date columns to datetime objects, extract date from time column, group heart rate data by ID and date, calculate the mean heart rate, and merge the dataframes.



```python
daily_activity_df['ActivityDate'] = pd.to_datetime(daily_activity_df['ActivityDate'])
heart_rate_df['Time'] = pd.to_datetime(heart_rate_df['Time'])
heart_rate_df['ActivityDate'] = heart_rate_df['Time'].dt.date
aggregated_heart_rate_df = heart_rate_df.groupby(['Id', 'ActivityDate'])['Value'].mean().reset_index()
aggregated_heart_rate_df.rename(columns={'Value': 'HeartRate'}, inplace=True)
aggregated_heart_rate_df['ActivityDate'] = pd.to_datetime(aggregated_heart_rate_df['ActivityDate'])

merged_df = pd.merge(daily_activity_df, aggregated_heart_rate_df, on=['Id', 'ActivityDate'])

display(merged_df.head())
```

## Prepare data

### Subtask:
Select the 'TotalSteps' and 'HeartRate' columns and handle any missing values.


**Reasoning**:
Select the specified columns, check for missing values, and remove rows with missing values.



```python
data = merged_df[['TotalSteps', 'HeartRate']].copy()

print("Missing values before dropping:")
print(data.isnull().sum())

data.dropna(inplace=True)

print("\nMissing values after dropping:")
print(data.isnull().sum())

display(data.head())
```

## Build and train model

### Subtask:
Build and train a Linear Regression model to predict heart rate from steps.


**Reasoning**:
Build and train a Linear Regression model to predict heart rate from steps using the prepared data.



```python
from sklearn.linear_model import LinearRegression

X = data['TotalSteps'].values.reshape(-1, 1)
y = data['HeartRate'].values

model = LinearRegression()
model.fit(X, y)
```

## Visualize the results

### Subtask:
Plot the heart rate against steps and add the regression line to the plot.


**Reasoning**:
Plot the scatter plot and the regression line as instructed.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalSteps', y='HeartRate', data=data)
sns.regplot(x='TotalSteps', y='HeartRate', data=data, scatter=False, color='red')
plt.title('Heart Rate vs. Total Steps with Regression Line')
plt.xlabel('Total Steps')
plt.ylabel('Heart Rate')
plt.show()
```
